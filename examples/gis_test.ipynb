```r
library("jupytergis")
```

```r
doc <- GISDocument$new("france_hiking.jGIS")
total_layers <- 3
doc
```

```r
google_layer <- doc$add_raster_layer(
    url = "https://mt1.google.com/vt/lyrs=y&x={x}&y={y}&z={z}",
    name = paste("Google Satellite", format(Sys.time(), "%H:%M:%S")),
    attribution = "Google",
    opacity = 0.6
)
total_layers <- total_layers + 1
total_layers
```

```r
# retrieve opacity
new_opacity <- doc$with_read(function(t) {
    doc$layers$get(t, google_layer)$parameters$opacity
})
new_opacity
```

```r
doc$with_read(function(t) {
    doc$layers$len(t)
    })
```

```r
assert_added <- function(doc=NULL, total=NULL) {
    layer_num <- doc$with_read(function(t) {
    doc$layers$len(t)
})

sources_num <- doc$with_read(function(t) {
    doc$sources$len(t)
})

stopifnot(
    layer_num == total,
    sources_num == total
    )
}
```

```r
assert_added(doc=doc, total=total_layers)
stopifnot(new_opacity == 0.6)
```

```r
osm_layer <- doc$add_raster_layer(
    url = "https://tile.openstreetmap.org/{z}/{x}/{y}.png",
    name = paste("OSM from R", format(Sys.time(), "%H:%M:%S")),
)
total_layers <- total_layers + 1
total_layers
```

```r
assert_added(doc=doc, total=total_layers)
```

```r
buildings <- doc$add_vectortile_layer(
    url = "https://planetarycomputer.microsoft.com/api/data/v1/vector/collections/ms-buildings/tilesets/global-footprints/tiles/{z}/{x}/{y}",
    name = paste("Buildings", format(Sys.time(), "%H:%M:%S")),
    attribution = "Planetary Computer",
    opacity = 1
)
total_layers <- total_layers + 1
total_layers
```

```r
assert_added(doc=doc, total=total_layers)
```

```r
roads <- doc$add_geojson_layer(
    path = "https://raw.githubusercontent.com/nvkelso/natural-earth-vector/master/geojson/ne_10m_roads.geojson"
)
total_layers <- total_layers + 1
total_layers
```

```r
# retrieve name
road_layer_name <- doc$with_read(function(t) {
    doc$layers$get(t, roads)$name
})
road_layer_name
```

```r
assert_added(doc=doc, total=total_layers)
stopifnot(road_layer_name == "ne_10m_roads")
```

```r
image <- doc$add_image_layer(
    url = "https://maplibre.org/maplibre-gl-js/docs/assets/radar.gif",
    coordinates = list(
    c(-80.425, 46.437),
    c(-71.516, 46.437),
    c(-71.516, 37.936),
    c(-80.425, 37.936)
  )
)
total_layers <- total_layers + 1
total_layers
```

```r
assert_added(doc=doc, total=total_layers)
```
